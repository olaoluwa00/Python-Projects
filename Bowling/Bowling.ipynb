```python
from tkinter import *
import numpy as np


root = Tk()

# Define labels for each bowling frame
# Define entry locations for each roll

Frame1Label = Label(root, text="Frame 1")
Frame1Label.grid(row=1, columnspan=2, column=0)

e1_1 = Entry(root, width=2)
e1_1.grid(row=2, column=0)

e1_2 = Entry(root, width=2)
e1_2.grid(row=2, column=1)

Frame2Label = Label(root, text="Frame 2")
Frame2Label.grid(row=1, columnspan=2, column=2)

e2_1 = Entry(root, width=2)
e2_1.grid(row=2, column=2)

e2_2 = Entry(root, width=2)
e2_2.grid(row=2, column=3)

Frame3Label = Label(root, text="Frame 3")
Frame3Label.grid(row=1, columnspan=2, column=4)

e3_1 = Entry(root, width=2)
e3_1.grid(row=2, column=4)

e3_2 = Entry(root, width=2)
e3_2.grid(row=2, column=5)

Frame4Label = Label(root, text="Frame 4")
Frame4Label.grid(row=1, columnspan=2, column=6)

e4_1 = Entry(root, width=2)
e4_1.grid(row=2, column=6)

e4_2 = Entry(root, width=2)
e4_2.grid(row=2, column=7)

Frame5Label = Label(root, text="Frame 5")
Frame5Label.grid(row=1, columnspan=2, column=8)

e5_1 = Entry(root, width=2)
e5_1.grid(row=2, column=8)

e5_2 = Entry(root, width=2)
e5_2.grid(row=2, column=9)

Frame6Label = Label(root, text="Frame 6")
Frame6Label.grid(row=1, columnspan=2, column=10)

e6_1 = Entry(root, width=2)
e6_1.grid(row=2, column=10)

e6_2 = Entry(root, width=2)
e6_2.grid(row=2, column=11)

Frame7Label = Label(root, text="Frame 7")
Frame7Label.grid(row=1, columnspan=2, column=12)

e7_1 = Entry(root, width=2)
e7_1.grid(row=2, column=12)

e7_2 = Entry(root, width=2)
e7_2.grid(row=2, column=13)

Frame8Label = Label(root, text="Frame 8")
Frame8Label.grid(row=1, columnspan=2, column=14)

e8_1 = Entry(root, width=2)
e8_1.grid(row=2, column=14)

e8_2 = Entry(root, width=2)
e8_2.grid(row=2, column=15)

Frame9Label = Label(root, text="Frame 9")
Frame9Label.grid(row=1, columnspan=2, column=16)

e9_1 = Entry(root, width=2)
e9_1.grid(row=2, column=16)

e9_2 = Entry(root, width=2)
e9_2.grid(row=2, column=17)

Frame10Label = Label(root, text="Frame 10")
Frame10Label.grid(row=1, columnspan=3, column=18)

e10_1 = Entry(root, width=2)
e10_1.grid(row=2, column=18)

e10_2 = Entry(root, width=2)
e10_2.grid(row=2, column=19)

e10_3 = Entry(root, width=2)
e10_3.grid(row=2, column=20)

# Convert values entered into list
def conc(a,b,c,d,e,f,g,h,i,j):
    List = [a, b, c, d, e, f, g, h, i, j]
            
    return (List)

# Convert list into String according to bowling rules
def String_Score(List):
    S = ""

    for item in List:
        if len(item) == 2:
            if item[0] == 10 and item[1] == 0:
                S += "X"
            elif item[0] + item[1] == 10:
                S += str(item[0]) + "/"
            elif item[0] + item[1] < 10:
                if item [0] > 0:
                    S += str(item[0])
                    if item[0] + item[1] == 10:
                        S += "/"
                    elif item[1] > 0:
                        S += str(item[1])
                    elif item[1] == 0:
                        S += "-"
                elif item[0] == 0:
                    S += "-"
                    if item[0] + item[1] == 10:
                        S += "/"
                    elif item[1] > 0:
                        S += str(item[1])
                    elif item[1] == 0:
                        S += "-"
        elif len(item) == 3:
            if item [0] == 10:
                S += "X"
                if item[1] == 10:
                    S += "X"
                    if item[2] == 10:
                        S += "X"
                    elif item[2] < 10:
                        S += str(item[2])
                elif item[1] < 10:
                    if item [1] > 0:
                        S += str(item[1])
                        if item[1] + item[2] == 10:
                            S += "/"
                        elif item[2] > 0:
                            S += str(item[2])
                        elif item[2] == 0:
                            S += "-"
                    elif item[1] == 0:
                        S += "-"
                        if item[1] + item[2] == 10:
                            S += "/"
                        elif item[2] > 0:
                            S += str(item[2])
                        elif item[2] == 0:
                            S += "-"
            elif item[0] < 10:
                if item[0] == 0:
                    S += "-"
                    if item[0] + item[1] == 10:
                        S += "/"
                        if item[2] == 10:
                            S += "X"
                        elif item[2] < 10:
                            S += str(item[2])
                    elif item[0] + item[1] < 10:
                        S += str(item[1])
                elif item[0] > 0:
                    S += str(item[0])
                    if item[0] + item[1] == 10:
                        S += "/"
                        if item[2] == 10:
                            S += "X"
                        elif item[2] < 10:
                            S += str(item[2])
                    elif item[0] + item[1] < 10:
                        if item[1] > 0:
                            S += str(item[1])
                        elif item[1] == 0:
                            S += "-"

    return S

# Convert list into array of scores based on bowling rules
def Score(List):
    n = len(List)
    Scores = np.zeros(n)

    for i in range (0,n-1):
        if List[i][0] == 10 and List[i][1] == 0:
            if List[i+1][0] == 10 and i < n-2:
                Scores[i] = List[i][0] + List[i+1][0] + List[i+2][0]
            else:
                Scores[i] = List[i][0] + List[i+1][0] + List[i+1][1]
        elif List[i][0] + List[i][1] == 10:
                Scores[i] = List[i][0] + List[i][1] + List[i+1][0]
        elif List[i][0] + List[i][1] < 10:
                Scores[i] = List[i][0] + List[i][1]
        else:
            print('ERROR in Frame: ' + str(i+1))
    Scores[n-1] = List[n-1][0] + List[n-1][1] + List[n-1][2]
    
    return Scores

# Iterate through array of to get final total game score
def Total_Score(Scores):
    Total = 0
    n = len(Scores)

    for i in range (n):
        Total += Scores[i]
    
    return Total

# Convert entries into integers from strings, and convert empty entries to zero
def get_integer(entrybox):
    if entrybox == "":
        return 0
    else:
        return int(entrybox)

# Store values when button is clicked
def GetScore():    

    Frame1 = np.array([get_integer(e1_1.get()), get_integer(e1_2.get())])
    Frame2 = np.array([get_integer(e2_1.get()), get_integer(e2_2.get())])
    Frame3 = np.array([get_integer(e3_1.get()), get_integer(e3_2.get())])
    Frame4 = np.array([get_integer(e4_1.get()), get_integer(e4_2.get())])
    Frame5 = np.array([get_integer(e5_1.get()), get_integer(e5_2.get())])
    Frame6 = np.array([get_integer(e6_1.get()), get_integer(e6_2.get())])
    Frame7 = np.array([get_integer(e7_1.get()), get_integer(e7_2.get())])
    Frame8 = np.array([get_integer(e8_1.get()), get_integer(e8_2.get())])
    Frame9 = np.array([get_integer(e9_1.get()), get_integer(e9_2.get())])
    Frame10 = np.array([get_integer(e10_1.get()), get_integer(e10_2.get()), get_integer(e10_3.get())])

    global OverallScore
    OverallScore = conc(Frame1, Frame2, Frame3, Frame4, Frame5, Frame6, Frame7, Frame8, Frame9, Frame10)

    LabelScore = Label(root, text= Total_Score(Score(OverallScore)))
    LabelScore.grid(row=4, columnspan=21)
    LabelString = Label(root, text= String_Score(OverallScore)) 
    LabelString.grid(row=5, columnspan=21)

# Define button to capture entry values
myButton = Button(root, text="Get Score", command=GetScore)
myButton.grid(row=3, columnspan=21)

root.mainloop()

print(OverallScore)
print(Score(OverallScore))
print(Total_Score(Score(OverallScore)))
print(String_Score(OverallScore))
```
